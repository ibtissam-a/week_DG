```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Data sets description 
In this mission we will use 4 individual csv fils that contains data collected from 31/01/2016 to 31/12/2018:
* Cab_Data.csv : it contains details about the transaction for 2 cab companies.
* City.csv : it gives us idea about the US cities, their population and the number of cab users.
* Customer_ID.csv : this is a mapping table that contains a unique identifier which links the customerâ€™s demographic details.
* Transaction_ID.csv : this is a mapping table that contains transaction to customer mapping and payment mode.

> *we  will study all those files one by one* 


# Cab_Data.csv

```python
# we load the first file then we dsiplay the first five rows 
d_Cab = pd.read_csv('../input/data-set-week-2/Cab_Data.csv')
d_Cab.head()
```

```python
print("The shape of the Cab data set is : ",d_Cab.shape)
```

```python
# check the columns 
d_Cab.columns 
```

***Numerical variables analysis**

```python
# display the numerical variables
d_Cab.select_dtypes(include=['number'])
```

> *so the numerical variables are: Transaction ID, Date of Travl, KM Traveled, Price charged, Cost of Trp*

```python
numerical_var = ['Transaction ID', 'Date of Travel', 'KM Travelled', 'Price Charged', 'Cost of Trip']
for var in numerical_var :
    plt.figure(figsize = (15,3))
    plt.hist(d_Cab[var] , bins = 50)
    plt.xlabel(var)
    plt.ylabel("Frequency")
    plt.title("The histogram of {} ".format(var))
    plt.show()
```

***Categorical variables analysis**

```python
# display the categorical variables
d_Cab.select_dtypes(include=['object'])
```

> *here we have only two categorical variables witch are : Company and City*

```python
categorical_var = ['Company', 'City']
for var in categorical_var:
    barplot = d_Cab[var].value_counts()
    plt.figure(figsize = (30,5))
    plt.bar(barplot.index, barplot)
    plt.ylabel("Frequency")
    plt.title(var)
    plt.show()
```

```python
d_Cab.describe()
```

```python
# detect missing values
d_Cab.isnull().sum()
```

```python
d_Cab.dtypes 
# we have to change the type of the company and city features
```

```python
dummies_Company = pd.get_dummies(d_Cab.Company)
dummies_Company
```

```python
dummies_City = pd.get_dummies(d_Cab.City)
dummies_City
```

```python
merged = pd.concat([d_Cab, dummies_Company], axis = 'columns')
merged = pd.concat([merged, dummies_City], axis = 'columns')
merged.head()
```

```python
final = merged.drop(['Company', 'City', 'Yellow Cab', 'PHOENIX AZ'], axis = 'columns')
final.head()
```

```python
final.columns 
```

```python
final.head()
```
