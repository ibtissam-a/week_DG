# Background Analysis

```python
# we merge the 4 data sets inorder to create a Master Data to be use in the fllowing steps 
Master_Data = d_Cab.merge(d_Transaction, on='Transaction ID').merge(d_Customer, on='Customer ID').merge(d_City, on='City')
Master_Data.head(5)
```

```python
# we have to eliminate the "," from the values of the two columns "Population" and "Users" inorder to facilitate the conversion
Master_Data['Population'] = Master_Data['Population'].str.replace(',', '')
Master_Data['Users'] = Master_Data['Users'].str.replace(',', '')
#d_City['Avg_Annual'] = df1['Avg_Annual'].astype(int)
```

```python
#convert the type of the 2 columns "Population" and "Users" from object type to int
Master_Data = Master_Data.astype({"Population": int, "Users": int})
```

```python
Master_Data['City'] = Master_Data['City'].astype('category')
Master_Data['Company'] = Master_Data['Company'].astype('category')
Master_Data['Payment_Mode'] = Master_Data['Payment_Mode'].astype('category')
Master_Data['Gender'] = Master_Data['Gender'].astype('category')
```

```python
quantitative_columns = Master_Data.select_dtypes('number')
for col in quantitative_columns:
    plt.figure()
    sns.distplot(Pink_df[col], label = 'Pink')
    sns.distplot(Yellow_df[col], label = 'Yellow')
    plt.legend()
```

```python
columns=['Transaction_ID', 'KM_Travelled', 'Price_Charged', 'Cost_of_Trip','Age','Population','Users']
Pink_df = Pink_df[columns]
Correlation_Pink = Pink_df.corr()
plt.figure(figsize=(20,10))
sns.heatmap(Correlation_Pink,annot=True,vmin=-1,vmax=1)
```

```python
columns=['Transaction_ID', 'KM_Travelled', 'Price_Charged', 'Cost_of_Trip','Age','Population','Users']
Yellow_df = Yellow_df[columns]
Correlation_Yellow = Yellow_df.corr()
plt.figure(figsize=(20,10))
sns.heatmap(Correlation_Yellow,annot=True,vmin=-1,vmax=1)
```
